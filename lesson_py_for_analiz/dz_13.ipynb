**Задание №13: Библиотека Seaborn**

**1 Дайте описание библиотеки Seaborn**

 Seaborn — библиотека для создания статистических графиков на Python. Она
построена на основе matplotlib и тесно интегрируется со структурами данных pandas.
Seaborn помогает вам изучить и понять данные. Его функции построения графиков
работают с датасетами и выполняют все необходимы преобразования для создания
информативных графиков

image.png

```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset('penguins')
sns.set_theme()
```

```python
penguins
```

**2 Как нарисовать гистограмму в Seaborn?**

Чтобы нарисовать гистограмму в Seaborn нужно вызвать функцию ...plot(), и передать
ей категориальные и числовые переменные, которые нужно визуализировать

```python
plt.figure(figsize=(6,6))
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='stack')
```

```python
plt.figure(figsize=(6,6))
sns.kdeplot(data=penguins, x='flipper_length_mm', hue='species', multiple='stack')
```

```python
#plt.figure(figsize=(13,13))
sns.displot(data=penguins, x='flipper_length_mm', hue='species', col='species')
```

```python


sns.displot(data=penguins, x='flipper_length_mm', hue='species',  col='species', kind = 'kde')
```

**3 Подумайте, что нужно сделать, чтобы построенная гистограмма стала горизонтальной?**

 Чтобы нарисовать горизонтальную, а не вертикальную гистограмму нужно просто
поменять местами переменные передаваемые в x и y


```python
sns.barplot(penguins,x = 'island', y = 'body_mass_g')
```

```python
sns.barplot(penguins, y = 'island', x = 'body_mass_g')
```

**4 Какие нужно применить параметры, чтобы задать цвет графика?**

Для этого нужно задать параметр color функции barplot и тогда цвет всех столбцов
изменится на заданный

```python
sns.barplot(penguins,x = 'island', y = 'body_mass_g', color = 'r')
```
