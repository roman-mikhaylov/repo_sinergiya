## 2. написать функцию для нахождения одинаковых элементов в двух списках

```python
sp1 = [1,2,3,4,5,6,77,888,9,0,"text"]
sp2 = [11,2,3,4,5,'text',77,9,0]
def my_func(sp1,sp2):
    result = []
    result = list(set(sp1) & set(sp2))
    return result
print(my_func(sp1, sp2))
```

двумерные списки

```python
n=4
```

```python
a=[[0]*n for i in range(n)]
a
```

```python
for i in range(n):
  for j in range(n):
    if i<j:
      a[i][j] = 0
    elif i>j:
      a[i][j] = 2
    else:
      a[i][j] = 1
for row in a:
   print(' '.join([str(elem) for elem in row]))
```

другой способ. заполним главную диагональ "1"

```python
for i in range(n):
  a[i][i]=1
a
```

заполним нулями все, что выше главной диагонали

```python
for i in range(n):
  for j in range(i+1,n):
    a[i][j] = 0
a
```

заполним двойуами все, что ниже главной диагонали

```python
for i in range(n):
  for j in range(0,i):
    a[i][j] = 2
a
```

```python
# компактная запись
n=4
a=[[0]*n for i in range(n)]
for i in range(n):
    a[i][i]=1
    for j in range(0,i):
       a[i][j] = 2
    for j in range(i+1,n):
       a[i][j] = 0
for row in a:
    print(' '.join([str(elem) for elem in row]))

```

```python
# через генератор
n=4
a=[0]*n
a=[[2]*i+[1]+[0]*(n-i-1) for i in range(n)]
for row in a:
    print(' '.join([str(elem) for elem in row]))
```
