```python
# по одному магазину
sales59 = sales[sales.shop_id == 59]
sales59.head()
```

```python
# проверяем дату на
sales59.date[40084]
```

```python
# переводим в time
ts = pd.to_datetime(sales59.date[40084])
ts
```

```python
print(ts.day)
print(ts.month)
print(ts.year)
print(ts.hour)
```

```python
sales59.date = pd.to_datetime(sales59.date)
# меняем в исходном датафрейме данные
sales59.set_index('date', inplace = True)
sales59
```

```python
# удалаяем столбцы
sales59 = sales59.drop(columns = ['date_block_num', 'item_id'])
sales59
```

```python
# считаем доходность 59 магазина и индекс меняем на дату
sales59['income'] = sales59['item_price'] * sales59['item_cnt_day']
sales59
```

```python
'''
если хотим получить данные по годам, готовим отдеьные датафреймы
'''
sales59_2013 = sales59[sales59.index.year == 2013]
sales59_2014 = sales59[sales59.index.year == 2014]
sales59_2015 = sales59[sales59.index.year == 2015]
```

```python
sales59_2014.income.sum() # данных нет
```

```python
'''
доход по годам
'''
print(f' 2013 : , {sales59_2013.income.sum()}')
print(f' 2014 : , {sales59_2014.income.sum()}')
print(f' 2015 : , {sales59_2015.income.sum()}')
```

```python
# среднемесячный доход ( лист)
month_income_2013 = [sales59_2013[sales59_2013.index.month == i].income.sum() for i in range(1,13)]
```

```python
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
month_income_2013 = pd.Series(month_income_2013, index = months)
month_income_2013.head(12)
```

```python
# средний доход
month_income_2013.mean()
```

```python
# самый прибыльный месяц
best_month_2013 =  month_income_2013.index[month_income_2013.argmax()]

best_month_2013
```

```python
sales59_2013.index.max()
```

```python
Dict = {'model': ['logistic regression', 'SVM', 'KNN', 'Decision Tree'], 'accuracy' : [0.96, None, None, 0.67]}
df = pd.DataFrame(Dict)
df
```

просто нулями

```python
df1 = df.fillna(0)
df1
```

минимальными значениями

```python
df2 = df.fillna(df['accuracy'].min())
df2
```

```python
# средним
df3 = df.fillna(df['accuracy'].mean())
df3
```

```python
# заполнить ближайшими значениями с низу
df4 = df.fillna(method='bfill')
df4
```

```python
# заполнить ближайшими значениями с верху
df5 = df.fillna(method='ffill')
df5
```
