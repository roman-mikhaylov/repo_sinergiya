```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.Series([1,2,3,4,5]) # свой формат -  пара индекс-значение + тип данных
s1
```

```python
'''
возврат в формат numpay
'''
s1.to_numpy()
```

```python
'''
индексация возможна не только через числа
'''
s2 = pd.Series([1,2,3,4,5],index =['a','b','c','d','e'])
s2
```

```python
# обащение по индексу
s2.iloc[3]
```

```python
s2['d']
```

```python
# создание мз словаря
d = {'a':2., 'b':3, 'c':4} # точка сразу переводит данные во float
s3 = pd.Series(d)
s3
```

```python
# вариаты индексации
print('s2[0] = ', s2.iloc[0])
print('s2[1] = ', s2['a'])
print('срез:', s2[2:4], '\n')
print('индексация условным выражением: s2[s2 <= 3] = ', s2[s2%2 == 1])
```

**табличные данные**

```python
# создание DataFrame

d= {'price': pd.Series([1,2,3]), 'count':pd.Series([10,20,30])}
df1 = pd.DataFrame(d)
df1
```

```python
# со своей индексацией
d= {'price': pd.Series([1,2,3,5], index = ['product1','product2', 'product3', 'product4']), 'count':pd.Series([10,20,30, 27], index = ['product1','product2', 'product3', 'product4'])}
df2 = pd.DataFrame(d)
df2
```

```python
df2.info()
```

```python
# список индексов и колонок
print(df2.index)
print(df2.columns)
```

```python
df2[df2.columns[1]]
```

```python
# создание из массива словарей
d3 = [{'price':2, 'count': 8}, {'price': 3, 'count': 10}]
df3 = pd.DataFrame(d3)
df3
```
