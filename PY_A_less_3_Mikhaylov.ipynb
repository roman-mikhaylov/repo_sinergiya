**1 Какие функции выполняет модуль Random?**

модуль предназначен для генерации массивов со случайными элементами

**2 Как создать массив с помощью модуля?**
Самый простой способ задать массив со случайными элементами — использовать
функцию sample (или random, или random_sample, или ranf — это всё одна и та же
функция).
Без аргументов возвращает просто число в промежутке [0, 1), с одним целым числом —
одномерный массив, с кортежем — массив с размерами, указанными в кортеже (все
числа — из промежутка [0, 1)).
С помощью функции randint или random_integers можно создать массив из целых чисел.
Аргументы: low, high, size: от какого, до какого числа (randint не включает в себя это
число, а random_integers включает), и size — размеры массива.
Также можно генерировать числа согласно различным распределениям (Гаусса, Парето и
другие). Чаще всего нужно равномерное распределение, которое можно получить с
помощь функции uniform.


```python
# из случайных целых чисел от 0 до N размером  m*n
import numpy as np
from numpy import random

N = 100
m = 4
n = 3
X = random.randint(N, size = (m,n))
print(X)
```

```python
# из случайных вещественных чисел от 0 до 1 размером  m*n
m = 4
n = 3
X = random.rand(m,n)
print(X)
```

```python
# из случайных элементов из заданного списка размером m*n
arr = [1,4,6,'ert', 456,6, -100]
s = random.choice(arr, size = (n,m))
print(s)
```

```python
# из готовых распределений
# нормальное рапределение (Гаусса)
from matplotlib import pyplot as plt
normal = random.normal(size=10000)
normal
```

```python
plt.figure(figsize = (12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
# равномерное распределение
uniform = random.uniform(size=1000000)
uniform
```

```python
plt.figure(figsize = (12,12))
plt.grid()
hist = plt.hist(uniform, bins=200)
```

```python
# треугольное распределение
triangular = random.triangular(0, 0.5, 1, size = 1000000)
triangular
```

```python
plt.figure(figsize = (12,12))
plt.grid()
hist = plt.hist(triangular, bins=200)
```

```python
# экспотенциальное
lambd = 1
exponential = random.exponential(lambd, size = 1000000)


plt.figure(figsize = (12,12))
plt.grid()
hist = plt.hist(exponential, bins=200)
```

```python
# рfспределение Парето (закон Ципфа)

k = 1000
pareto = random.pareto(k, size = 10000000)


plt.figure(figsize = (12,12))
plt.grid()
hist = plt.hist(pareto, bins=200)
```

**3 Назовите способ сделать случайную выборку из массива?**

```python
# исходный массив
A = np.arange(0,10)
A
```

```python
# вариант без изменения исходного массива
A_1 = random.permutation(A)
A_1
```

```python
# с изменениями в исходном массиве
B = A.copy()
random.shuffle(B)
B
```

```python
# если нужно выбрать size=18 элементов из [colors] с вероятностями P(1/3, 1/6, 0, 1/2, 0)
colors=['желтый', 'зеленый', 'синий', 'красный', 'оранжевый']
x = random.choice(colors, size=(18), p=[1/3, 1/6, 0, 1/2, 0])
x
```

**4 Что вообще такое - случайные числа?**

Случайные числа представляют собой произвольные данные, которые были получены в
результате автоматической генерации компьютерной программой
Однако, строго говоря, эта генерация не является полностью случайной - в основе лежит функция или алгоритм.

"Основная сложность генерации последовательности псевдослучайных чисел на компьютере в том, что компьютеры детерминистичны по своей сути. Компьютер может находиться только в конечном количестве состояний (количество состояний огромно, но все-таки конечно). Следовательно, любой датчик случайных чисел по определению периодичен. Все периодическое — предсказуемо, т. е. не случайно."
