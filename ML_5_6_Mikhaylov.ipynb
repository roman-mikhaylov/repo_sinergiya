**Обьединение потребления энергии и погоды**



выставим индексы  для обьединения timstamp, sity_id

```python
energy_0.set_index(['timestamp', 'site_id'], inplace=True)
weather.set_index(['timestamp', 'site_id'], inplace=True)
```

проводим обьедиение и сбросим индексы

```python
energy_0 = pd.merge(left=energy_0, right=weather, how = 'left', left_index = True, right_index =	True)
energy_0.reset_index(inplace=True)
energy_0.head()
```

**2. вычислите первую производную для температуры воздуха и записать ее в столбец air_temperature1**

```python
# Вычисляем производную для столбца
from sympy import *
#energy_0['air_temperature1'] = energy_0['air_temperature'].apply(lambda x: np.diff(x)/np.diff(x.index))

energy_0['air_temperature1'] = energy_0['air_temperature'].diff() # первая будет Nan, потом удалим строку
energy_0
```

**3 вычислите синус и косинус от угла направления ветра, записать их в столбцы wind_direction_cos, wind_direction_sin**

```python
energy_0['wind_direction_cos'] = np.cos(energy_0['wind_direction'])
energy_0['wind_direction_sin'] = np.sin(energy_0['wind_direction'])
energy_0
```

**4.Удалить столбцы site_id building_id meter	primary_use square_feet year_built floor_count wind_direction**

```python
energy_0.drop(['site_id', 'building_id', 'meter', 'primary_use', 'square_feet', 'year_built', 'floor_count', 'wind_direction'], axis=1, inplace=True)
energy_0.info()
```

**5. Удалите из данных строки с metr_reading = 0**

```python
#df = df[df['col']] != 'target'

index_names = energy_0[energy_0['meter_reading'] == 0].index
energy_0.drop(index_names, inplace = True)
energy_0
```
