**1.Что такое парсинг данных и его значение в ML?**

Парсинг данных - это процесс сбора и обработки информации с целью извлечения значимых данных. В контексте машинного обучения (ML) парсинг играет важную роль, поскольку позволяет получить качественные и релевантные данные для обучения моделей.

Значение парсинга данных в ML:

1.Получение необходимых данных: Парсинг позволяет собирать большие объемы данных из различных источников, таких как веб-сайты, базы данных или социальные сети. Это обеспечивает доступ к большим объемам информации, необходимой для обучения ML-моделей.

2.Фильтрация и очистка данных: Парсеры могут автоматически фильтровать и очищать полученные данные, удаляя нежелательные значения, пропуски или дубликаты. Это помогает улучшить качество данных, что в свою очередь повышает точность и эффективность моделей машинного обучения.

3.Разделение данных на обучающую, валидационную и тестовую выборки: Парсинг может помочь разделить данные на различные выборки для разных этапов обучения и оценки модели. Это необходимо для обеспечения точности и надежности результатов.
Извлечение признаков и признаков: Парсеры также могут извлекать из данных определенные признаки или атрибуты, которые могут быть полезны для обучения модели. Например, это могут быть определенные ключевые слова, характеристики продукта или характеристики пользователя.
   
4.Ускорение процесса машинного обучения: Благодаря автоматизации процессов сбора, фильтрации и разделения данных, парсинг помогает ускорить процесс машинного обучения, позволяя сосредоточиться на более сложных задачах и анализе данных.

**2.В чём сильные стороны лямбда архитектуры и её сложность для реализации?**

Лямбда-архитектура в ML (машинном обучении) предлагает ряд преимуществ, таких как гибкость, масштабируемость и отказоустойчивость. Однако ее реализация может быть сложной из-за необходимости разработки микросервисов и учета различных факторов, таких как производительность и безопасность, а также необходимость на выходе правильно совмещать два обработанных потока (пакетного и потокового)

Одним из преимуществ лямбда-архитектуры в ML является ее гибкость. Благодаря использованию микросервисов, можно легко добавлять новые модели, обновлять существующие и интегрировать внешние источники данных. Это позволяет быстрее реагировать на изменение требований и адаптировать систему к новым задачам.

Еще одним преимуществом является масштабируемость. Каждый микросервис может быть масштабирован отдельно, что позволяет системе адаптироваться к изменениям в нагрузке или требованиям без прерывания работы.

Однако реализация лямбда-архитектуры в ML может быть сложной. Разработка микросервисов требует глубоких знаний в области машинного обучения, программирования и DevOps. Нужно учитывать факторы, такие как производительность моделей, безопасность данных и интеграция с другими системами.

Кроме того, важно учитывать возможность отказоустойчивости. Лямбда-архитектура обеспечивает высокую степень отказоустойчивости за счет распределения нагрузки между микросервисами, но это также требует тщательного планирования и реализации.
