```python
import numpy as np
import pandas as pd
```

```python
# жанры
g = {1:'darama', 2:'action', 3:'Sci-Fi', 4 :'horor'}
```

```python
# исходный список
array = np.array([['Interstellar', g[3], 2014, 12300., 4.2],
               ['The Shawshank Redemption', g[1], 1994, 630., 3.5],
               ['Matrix-1', g[3], 2001, 10001., 3.6],
                ['Matrix-2', g[3], 2003, 5000., 4.7],
                ['Matrix-3', g[3], 2011, 80000., 3.7],
                ['кошмар на улице вязов -1', g[4], 1987, 20000., 5.0],
                ['звонок -1', g[4], 1994, 1000., 5.],
                ['звонок -2', g[4], 2000, 20000., 3.4],
                ['терминатор -1', g[2], 1987, 5000., 5.],
                ['Matrix-8', g[3], 2035, 80000., -4.6]])
array
```

```python
# создаем датафрейм
df = pd.DataFrame(array, columns=['film','genre', 'year', 'quantity', 'estimation'])
df
```

```python
# типы данных
print(df.dtypes)
```

```python
df.info()
```

```python
# меняем тип данных для "quantity"
df.quantity = df.quantity.astype(float)
df.info()
```

1.Вывести только фильмы с определённым жанром, например, "хоррор"

```python
df1 = df.loc[df['genre'] == 'horor']
df1
```

2.Вывести только фильмы, которые были сняты после 2010 года

```python
df2 = df.loc[df['year'] > '2010']
df2
```

3.Вывести только фильмы с оценкой выше 4.0

```python
df3 = df.loc[df['estimation'] > '4.0']
df3
```

4.Вывести только фильмы, у которых больше 10 000 оценок

```python
df4 = df.loc[df['quantity'] > 10000]
df4
```
