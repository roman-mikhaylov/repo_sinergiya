```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,5,3,4.5]) # ломанную рисует 1,3 - индексы по X оси, 5,4.5 по Y
plt.ylabel('some numbers')
plt.show()
```

**2 Определения Figure, Axes, Spine**

Всё пространство (figure) включает в себя график (Axes), который, в свою очередь, состоит из таких элементов, как границы (Spines),
сетки (Grid), описания (Legend), самого графика (Line) и других.
Графики отражают зависимость, при этом элементы графика — оси, подписи и др. — открыты к настройке.
Базовый элемент figure может содержать в себе как один график, так и несколько.

image.png

```python
# на примере, несколько графиков на однои Figure
plt.figure(figsize=(6,7)) # пространство графиков размером 6*7
plt.plot([1,2,3,4,5],[5, 6, 4.5,5, 10], "go")
plt.plot([1,2,3,4,5],[5, 6, 4.5,5, 10], "r--")
plt.axis([0,6,0,20]) # 0-6 ось X, 0-20 ось Y, двумерная плоскость

t = np.arange(0.,5.,0.2)
plt.plot(t,t,'bs--', t,t**2, 'g',t,t**3, 'r^')


plt.show()

```

```python
# координатная секта (Grid)
plt.figure(figsize=(6,7))

plt.grid()
plt.plot(t,t,'bs--', t,t**2, 'g--',t,t**3, 'r^--')

plt.show()
```

```python
# подписи (Anacomy of a figure)
plt.figure()

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Function:$f(x) = x, f(x) = x^2$')

plt.grid()
plt.plot(t,t,'bs--', t,t**2, 'g--',t,t**3, 'r^--')

plt.show()
```

```python
# и несколько графиков
def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize=(10,10))
plt.subplot(211)
plt.grid(True)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.grid(True)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
