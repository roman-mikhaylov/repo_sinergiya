```python
import math

x = np.arange(0,math.pi*9,0.1)
y = np.cos(x)
y1 = np.sin(x)
y2 = 0 * x + 0.5

plt.plot(x,y,x,y1,'b.',x,y2, 'r-')
plt.xlabel("частота")
plt.ylabel("амплитуда")
plt.title("cos")
plt.show()
```

```python
x = np.arange(10)
y = [a **2 for a in range(10)]
y2 = np.arange(10)+20
y3 = [a **2 for a in range(10,0,-1)]


plt.plot(x,y, 'ro-.', x, y2, x, y3, 'bo-.')
plt.legend(labels=('square','babah', 'reserved square'))
```

```python
# случайные цветные значения
fig, ax = plt.subplots(figsize = (16,9))
ax.plot(np.sin(np.random.randint(10, size=100)),'o-')
ax.plot(np.cos(np.random.randint(10, size=100)),'dr-')
ax.plot(np.tan(np.random.randint(10, size=100)),'xg-')
ax.plot(np.random.randint(100, size=100),'ok-')
ax.plot(np.random.randint(100, size=100),'dr--')
ax.plot(np.random.randint(100, size=100),'o-.')
```
