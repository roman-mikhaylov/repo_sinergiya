**1.В чём отличие ML от классического программирования?**

Machine Learning (ML) и классическое программирование имеют ряд отличий. Вот некоторые из них:

1.Цель: Цель классического программирования заключается в создании алгоритмов и программ, которые выполняют предопределенные задачи. В то время как цель ML - это создание алгоритмов и моделей, которые могут автоматически извлекать и анализировать данные для выполнения задач без явного программирования.

2.Обучение: Классические программы обычно требуют написания кода для определения входных и выходных данных. В отличие от этого, модели ML обучаются на больших наборах данных, используя алгоритмы обучения, такие как машинное обучение, глубокое обучение и другие.

3.Эффективность: Классическое программирование может быть более эффективным по сравнению с ML в случаях, когда алгоритм или процесс уже известны и определены. Однако ML может быть эффективнее, если требуется обработка больших объемов данных или если алгоритм или процесс еще не определены.

4.Масштабируемость: ML алгоритмы могут быть более масштабируемыми, чем классические программы, так как они способны адаптироваться к новым данным и задачам. Классические программы могут быть менее масштабируемыми в некоторых случаях.

5.Человеческий фактор: Классическое программирование требует написания кода и разработки алгоритмов, что требует знаний программирования и технических навыков.
ML же может быть более доступным для нетехнических специалистов, поскольку оно позволяет использовать существующие алгоритмы и инструменты для решения задач.

6.Точность: Классические алгоритмы могут быть менее точными по сравнению с ML, так как последние могут использовать статистические и вероятностные методы для повышения точности результатов.

В целом, ML и классическое программирование дополняют друг друга и могут использоваться в зависимости от конкретной задачи и требований.

```python
import pandas as pd
import numpy as np
```

**очистка данных ETL**

```python
xls = pd.ExcelFile("https://video.ittensive.com/math-stat/demo13.xls") # получаем XLS файл
```

```python
sheet = xls.parse(0) # берем 1 лист
sheet
```

1. выбираем 4 столбец (Unnamed: 3)
2. берем строки начиная с 6 до 34
3. сбрасываем индекс
4. выбираем нужный столбец
5. преобразовываем серию данных (женщины) во фрейм

```python
data  = pd.DataFrame({"женщины": sheet['Unnamed: 3'][6:34].reset_index()["Unnamed: 3"]})
```

```python
data.head()
```

**Построение и обучение модели**

возьмем среднее как самую простую модель

```python
model = float(data.mean())
print('{0:.4}'.format(model))
```

**вычисляем точночть модели**

рассчитаем ошибку по тестовым данным

```python
err = np.abs(data - model).mean()
print('{0:.3}%'.format(float(100*err/model)))
```

**применение модели**

сравним численность женского населения России 2021 с моделью

```python
# 2021 - 78,9 млн
result = 78.9
print('{0:.3}%'.format(100*(result - model)/result))
```
